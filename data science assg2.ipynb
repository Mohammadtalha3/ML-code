```python
import pandas
dt= pandas.read_csv('sales_data.csv')
col=dt.columns
dt['Total Sales']= dt[col[1:]].sum(axis=1)
dt['Min Sales']= dt[col[1:]].min(axis=1)
dt['Max Sales']= dt[col[1:]].max(axis=1)
print('Sales Report')
dt
```

```python
data=dt[col[1:]].transpose()
Data={}
Data['Max Sales']=data.max(axis=1)
Data['Min Sales']= data.min(axis=1)
Data['Avg Sales']=data.mean(axis=1).round(2)
Data['Total Sales']= data.sum(axis=1)

d= pandas.DataFrame(Data)
print('Products Statistics')
d

```

```python
import matplotlib.pyplot as pyp
for col in ['ProductA','ProductB','ProductC','ProductE','ProductF']:
    pyp.plot(dt['Day'], dt[col])
    pyp.title(col)
    pyp.ylabel('Sales')
    pyp.show()



```
