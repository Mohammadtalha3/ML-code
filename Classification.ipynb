```python
import pandas
dt= pandas.read_csv('Rainfall_data.csv')

```

```python
from sklearn.model_selection import train_test_split 
inp=dt.drop(columns=['Index', 'Rain'])
out=dt['Rain']
inp_x,tst_x,out_y,tst_y=train_test_split(inp,out,test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeClassifier
CModel= DecisionTreeClassifier()
CModel.fit(inp_x, out_y)
```

```python
pred_y= CModel.predict(tst_x)
```

```python
com= pandas.DataFrame({'Actual Values':tst_y.values,
                      'Predicted Values':pred_y.flatten()})
```

```python
from sklearn.preprocessing import LabelEncoder
lEnc=LabelEncoder()
lEnc.fit(['Yes','No'])
          
```

```python
com['Actual Values'] = lEnc.transform(com['Actual Values'])
com['Predicted Values']= lEnc.transform(com['Predicted Values'])


```

```python
import matplotlib
com.plot(kind='density')
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(tst_y,pred_y)*100
print('Accuracy of the model:',acc,'%')
```
