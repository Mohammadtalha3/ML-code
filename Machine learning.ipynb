```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas
dt=pandas.read_csv('height_and_weight.csv')
```

```python
dt.drop(columns=['Index'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
height=dt.drop(columns=['Weight(lbs)'])
weight=dt.drop(columns=['Height(In)'])
# #training set
# inp_x= weight.values[:180]
# out_y= height.values[:180]
#  #testing set
# tst_x= weight.values[180:]
# tst_y= height.values[180:]
inp_X,tst_X,out_y,tst_y=train_test_split(height,weight, test_size=0.1)

```

```python
Model=LinearRegression()
Model.fit(inp_X,out_y)
```

```python
pred_y =Model.predict(tst_X)
```

```python
com= pandas.DataFrame({'Predicted Values' : pred_y.flatten(),
                     'Actual Values': tst_y['Weight(lbs)'].values})
                     
com.plot.bar(figsize=(10,6))
```

```python
from matplotlib import pyplot
pyplot.scatter(tst_X,tst_y)
pyplot.plot(tst_X['Height(In)'],pred_y)
```
