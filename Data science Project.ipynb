```python

import pandas as pd
data=pd.read_csv('trees.csv',
                names=['I','Girth','Height','Volume'])
#cleaning Data
del data['I']
data.drop(0, inplace =True)

data['Girth']= data['Girth'].astype(float)
data['Height']= data['Height'].astype(float)
data['Volume']=data['Volume']. astype(float)

data['Girth']*=0.0254
data['Height']*=0.305
data['Volume']*=0.0283

#we have to round the data again because when we access the the calculations directly
```

```python
#Data Analysis
max_girth= (data['Girth'].round(2).max())
max_height= (data['Height'].round(1).max())
max_volume=(data['Volume'].round(1).max())


min_girth=(data['Girth'].round(2).min())
min_height=(data['Height'].round(1).min())
min_volume=(data['Volume'].round(1).min())

avg_girth=round(data['Girth'].mean(),2)
avg_height=round(data['Height'].mean(),2)
avg_volume= round(data['Volume'].mean(),2)


```
