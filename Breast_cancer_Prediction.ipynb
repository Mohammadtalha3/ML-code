```python
import pandas
dt=pandas.read_csv('breast-cancer.csv')
```

```python
dt.columns
```

```python
#In this dataset we can use 3 types of algorithms Random forest, logistic regression, KNN, test this program to all and test 
#the accuracy of all 
```

```python
from sklearn.model_selection import train_test_split
x= dt.drop(columns=['diagnosis'])
y=dt['diagnosis']
inp_x,tst_x,out_y,tst_y=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(inp_x,out_y)
```

```python
pred_y=model.predict(tst_x)
pred_y
```

```python
from sklearn.metrics import accuracy_score
pd=accuracy_score(tst_y,pred_y)*100
pd
```

```python
from sklearn import metrics
mt= metrics.confusion_matrix(tst_y,pred_y)
mt
```
